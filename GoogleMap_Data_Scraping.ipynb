```python


cities = ["NL Complex", "Mandapeshwar Caves", "Northern heights", "Shakti Nagar", "Anand Nagar", "Ketkipada", "Anand Park", "Krishna Colony", "Rawalpada", "Ashok Van", "Balaji Colony", "Ekta Colony", "Maratha Nagar",
"Konkani pada","CS Complex","Avdhut Nagar","Narendra Complex","Shanti Nagar","Yadav Nagar","Gavde Nagar","Ganpat Patil Marg","Navagaon", "Mhatre Wadi", "Tadwe Wadi"]
```

```python
Name = []
Address = []
Number = []

def searchPlate():
    ## locating the search bar
    plate = driver.find_element(By.XPATH, "/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input")
    # plate = driver.find_element(By.XPATH,"/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input")
    sleep(3)
    ## looping through different cities
    for idx, val in enumerate(cities):
        # plate.send_keys(f"Construction company in {val} andheri")
        plate.send_keys(f"Construction companies in {val} Dahisar")
        print("Scraping Construction cos in", val)
        sleep(3)
        ## Locating the search button and clicking it
        submit = driver.find_element(By.XPATH, "/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button")
        # submit = driver.find_element(By.XPATH, "/html/body/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button")
        submit.click()
        sleep(20)
        
        # driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        cards = driver.find_elements(By.CLASS_NAME, "Nv2PK")
        sleep(30)
        ## Scraping data from the cards and appending it to an empty list
        for c in cards:
            name = c.text.split("\n")[0]
            add = c.text.split("\n")[2]
            num = c.text.split("\n")[3]
            Name.append(name)
            Address.append(add)
            Number.append(num)
        ## clear name of the place
        plate.clear()
searchPlate()
```

```python
df = pd.DataFrame({"Name" : Name, "Address" : Address, "Number" : Number})
```
